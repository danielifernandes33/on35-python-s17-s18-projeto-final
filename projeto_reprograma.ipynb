###Filtrando o DataFrame (df1)

```python
# Eliminar as colunas de idade, altura, peso e equipe
# axis=1 indica que você está trabalhando com colunas
df1 = df1.drop(['equipe', 'idade','altura', 'peso', 'delegacao'], axis=1)
```

```python
# Substituindo o Nan para 0 na coluna medalha (esta coliuna informa se o atleta medalha de ouro, prata, bronze ou não ganhou medalha)
df1.loc[:, 'medalha'] = df1['medalha'].fillna(0)
```

```python
# Criação da coluna "medalha_atleta" com 1 para ganhou medalha e 0 para não ganhou medalha
df1['medalha_atleta'] = df1['medalha'].apply(lambda x: 0 if x == 0 else 1)
```

##Visualizando Dados

```python
# cria um gráfico de linhas usando a biblioteca matplotlib
# eixo x (horizontal): Representa os anos.
# eixo y (vertical): Representa o número de atletas
contagem_atletas_por_ano = df1.groupby('ano').size()

plt.figure(figsize=(10, 6))
plt.plot(contagem_atletas_por_ano.index, contagem_atletas_por_ano.values, marker='o', linestyle='-', color='g')

plt.title('Participação do Brasil em Olímpiadas \n1900 a 2016')
plt.xlabel('Ano')
plt.ylabel('Número de Atletas')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
# Agrupar a contagem de atletas por ano
contagem_atletas_por_ano = df1.groupby('ano').size()

# Configurar a figura e o tamanho
plt.figure(figsize=(14, 6))

# Criar o gráfico de barras
plt.bar(contagem_atletas_por_ano.index, contagem_atletas_por_ano.values, color='g')

# Adicionar título e rótulos
plt.title('Participação do Brasil em Olímpiadas \n1900 a 2016')
plt.xlabel('Ano')
plt.ylabel('Número de Atletas')

# Definir os ticks do eixo x para mostrar todos os anos
plt.xticks(contagem_atletas_por_ano.index, rotation=45)

# Adicionar grade e ajustar layout
plt.grid(axis='y')
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
# Transformar o DataFrame para o formato longo para auxiliar a criação dos gráficos
# Agrupar o DataFrame original 'df' pelos campos 'ano' e 'sexo'
# Aplicar uma função de agregação para contar o número de atletas.agg(
# Criar uma nova coluna chamada 'contagem_atletas' que contará o número de 'id_atleta' para cada combinação de 'ano' e 'sexo'
# Resetar o índice para transformar o resultado em um DataFrame regular
# que pode ser manipulado facilmente e não terá mais um índice hierárquico

df_longo = df1.groupby(['ano', 'sexo']).agg(contagem_atletas=('id_atleta', 'count')).reset_index()

# Exibir o DataFrame no formato longo
print(df_longo)
```

```python
# Gerar o gráfico de barras agrupadas (matplotlib)
plt.figure(figsize=(14, 6))

# Usar 'bar' e pivotar para garantir que os dados estejam na estrutura certa
# Especifica as cores para as barras
df_longo.groupby(['ano', 'sexo'])['contagem_atletas'].sum().unstack().plot(kind='bar',
    ax=plt.gca(), color=['green', 'blue'])  # Define as cores verde e amarelo

# Adicionar título e rótulos
plt.title('Comparação dos Atletas por Gênero entre os anos \n 1900 a 2016')
plt.xlabel('Ano')
plt.ylabel('Número de Atletas')
plt.xticks(rotation=45)

# Adicionar legenda
plt.legend(title='Sexo', labels=['Feminino', 'Masculino'])

# Adicionar grade e ajustar layout
plt.grid(True)
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
# Gerar o gráfico de linhas para atletas femininos e masculinos por ano (biblioteca seaborn)
plt.figure(figsize=(14, 6))

# Usar 'line' e pivotar para garantir que os dados estejam na estrutura certa
# Especifica as cores e a espessura das linhas
df_longo.groupby(['ano', 'sexo'])['contagem_atletas'].sum().unstack().plot(kind='line',
    ax=plt.gca(), color=['green', 'blue'], linewidth=2.5)  # Aumenta a espessura das linhas

# Adicionar título e rótulos
plt.title('Comparação dos Atletas por Gênero entre os anos \n 1900 a 2016')
plt.xlabel('Ano')
plt.ylabel('Número de Atletas')

# Definir todos os anos no eixo x
plt.xticks(ticks=df_longo['ano'].unique(), rotation=45)  # Ajuste para rotacionar os rótulos

# Adicionar legenda
plt.legend(title='Sexo', labels=['Feminino', 'Masculino'])

# Adicionar grade e ajustar layout
plt.grid(True)
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
# Filtrando os eventos coletivos e únicos
df_coletivos_unicos = df1[df1['individual_coletivo'] == 'Coletivo'].drop_duplicates(subset=['ano', 'evento'])

# Contar as medalhas individuais
medalhas_individuais = df1[df1['individual_coletivo'] == 'Individual'].groupby('ano')['medalha_atleta'].sum().reset_index(name='medalhas_individuais')

# Contar as medalhas coletivas (apenas uma por evento)
medalhas_coletivas = df_coletivos_unicos.groupby('ano')['medalha_atleta'].sum().reset_index(name='medalhas_coletivas')

# Juntar as contagens de medalhas individuais e coletivas
medalhas_por_ano = pd.merge(medalhas_individuais, medalhas_coletivas, on='ano', how='outer')

# Preencher NaN com 0 para medalhas não existentes
medalhas_por_ano.fillna(0, inplace=True)

# Converter as colunas para o tipo inteiro
medalhas_por_ano['medalhas_individuais'] = medalhas_por_ano['medalhas_individuais'].astype(int)
medalhas_por_ano['medalhas_coletivas'] = medalhas_por_ano['medalhas_coletivas'].astype(int)

# Somar as medalhas coletivas e individuais
medalhas_por_ano['total_medalhas'] = medalhas_por_ano['medalhas_individuais'] + medalhas_por_ano['medalhas_coletivas']

# Exibir o DataFrame resultante
print(medalhas_por_ano)
```

```python
# Criar o gráfico
plt.figure(figsize=(12, 6))

# Criar um gráfico de barras para total de medalhas
plt.bar(medalhas_por_ano['ano'], medalhas_por_ano['total_medalhas'], color='green')

# Adicionar título e rótulos
plt.title('Total de Medalhas Conquistadas por Ano')
plt.xlabel('Ano')
plt.ylabel('Número Total de Medalhas')

# Definir ticks no eixo x para todos os anos
plt.xticks(medalhas_por_ano['ano'], rotation=45)  # Exibir todos os anos no eixo x com rotação de 45 graus

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

```python
# Criar o gráfico de linhas
plt.figure(figsize=(12, 6))

# Gráfico de linhas para o total de medalhas
plt.plot(medalhas_por_ano['ano'], medalhas_por_ano['total_medalhas'], marker='o', color='green', linestyle='-', linewidth=2, markersize=5)

# Adicionar título e rótulos
plt.title('Total de Medalhas Conquistadas por Ano')
plt.xlabel('Ano')
plt.ylabel('Número Total de Medalhas')

# Definir ticks no eixo x para todos os anos
plt.xticks(medalhas_por_ano['ano'], rotation=45)  # Exibir todos os anos no eixo x com rotação de 45 graus

# Exibir a grade
plt.grid()

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

```python
# Filtrar os eventos coletivos e individuais
df_coletivos_unicos = df1[df1['individual_coletivo'] == 'Coletivo'].drop_duplicates(subset=['ano', 'evento'])

# Contar as medalhas individuais
medalhas_individuais = df1[df1['individual_coletivo'] == 'Individual'].groupby(['esporte', 'ano'])['medalha_atleta'].sum().reset_index(name='medalhas_individuais')

# Contar as medalhas coletivas (apenas uma por evento)
medalhas_coletivas = df_coletivos_unicos.groupby(['esporte', 'ano'])['medalha_atleta'].sum().reset_index(name='medalhas_coletivas')

# Juntar as contagens de medalhas individuais e coletivas
medalhas_por_esporte = pd.merge(medalhas_individuais, medalhas_coletivas, on=['esporte', 'ano'], how='outer')

# Preencher NaN com 0 para medalhas não existentes
medalhas_por_esporte.fillna(0, inplace=True)

# Converter as colunas para o tipo inteiro
medalhas_por_esporte['medalhas_individuais'] = medalhas_por_esporte['medalhas_individuais'].astype(int)
medalhas_por_esporte['medalhas_coletivas'] = medalhas_por_esporte['medalhas_coletivas'].astype(int)

# Somar as medalhas coletivas e individuais por esporte
medalhas_por_esporte['total_medalhas'] = medalhas_por_esporte['medalhas_individuais'] + medalhas_por_esporte['medalhas_coletivas']

# Agrupar por esporte e somar as medalhas
total_medalhas_por_esporte = medalhas_por_esporte.groupby('esporte')['total_medalhas'].sum().reset_index()

# Classificar os esportes pela quantidade total de medalhas
total_medalhas_por_esporte = total_medalhas_por_esporte.sort_values(by='total_medalhas', ascending=False)

# Exibir o DataFrame resultante
print(total_medalhas_por_esporte)
```

```python
# Dicionário para traduzir os esportes

traducao_esportes = {
    'Athletics': 'Atletismo',
    'Swimming': 'Natação',
    'Football': 'Futebol',
    'Volleyball': 'Vôlei',
    'Basketball': 'Basquete',
    'Judo': 'Judô',
    'Sailing': 'Vela',
    'Beach Volleyball': 'Vôlei de Praia',
    'Archery': 'Arremesso de Arco',
    'Boxing': 'Boxe',
    'Shooting': 'Tiro Esportivo',
    'Gymnastics': 'Ginásttica',
    'Equestrianism': 'Hipismo',
    'Cycling': 'Ciclismo',
    'Canoeing': 'Canoagem',
    'Taekwondo': 'Taekwondo',
    'Modern Pentathlon': 'Pentatlo Moderno',

}

# Aplicando a tradução nos nomes dos esportes
total_medalhas_por_esporte['esporte'] = total_medalhas_por_esporte['esporte'].replace(traducao_esportes)
```

```python
# Filtrar esportes que trouxeram medalhas
total_medalhas_por_esporte = total_medalhas_por_esporte[total_medalhas_por_esporte['total_medalhas'] > 0]

# Ordenar os dados do menor para o maior (para inverter a exibição)
total_medalhas_por_esporte = total_medalhas_por_esporte.sort_values(by='total_medalhas', ascending=True)

# Definindo o tamanho do gráfico
plt.figure(figsize=(14, 6))

# Criando o gráfico de colunas horizontais
plt.barh(total_medalhas_por_esporte['esporte'], total_medalhas_por_esporte['total_medalhas'], color='green')

# Adicionando título e rótulos
plt.title('Total de Medalhas por Esporte nos\n Jogos Olímpicos de 1900 a 2016', fontsize=14)
plt.xlabel('Total de Medalhas', fontsize=12)
plt.ylabel('Esporte', fontsize=12)

# Exibindo os valores no gráfico
for index, value in enumerate(total_medalhas_por_esporte['total_medalhas']):
    plt.text(value, index, str(value))

# Exibindo o gráfico
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
